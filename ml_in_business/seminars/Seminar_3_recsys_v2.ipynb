## 0. Метрики

### 0.1. Precision@K


\begin{align}
    \text{Precision@k} = \frac{\text{Число релевантных items в топ-} k \text{ рекомендациях}}{k}
 \end{align}

\begin{align}
    \text{Важно!}
 \end{align}
 \begin{align}
    \text{Делим на } k \text{ даже если кандидатов сгенерировалось меньше}
 \end{align}



### 0.2. Recall@K

\begin{align}
    \text{Recall@k} = \frac{\text{Число релевантных items в топ-} k \text{ ркомендаций}}{\text{Всего релевантных items}}
 \end{align}

### 0.3. AP@K

\begin{align}
\text{AP@k} = \frac{1}{\text{Число релевантных items}} \sum_{i=1}^{k} \text{Precision@}i \times \text{Rel}_i
\end{align}

\begin{align}
\text{Rel}_i = 1
\text{ для релевантных элементов и 0 для нерелевантных, }
\end{align}

\begin{align}
\text{ если уровней релевантности больше, используем другие метрики}
 \end{align}



### 0.4. MAP@K

\begin{align}
\text{MAP@k} = \frac{1}{|U|} \sum_{u=1}^{|U|} \text{AP@k}_u
\end{align}

\begin{align}
\text{|U| } - \text{число пользователей}
\end{align}

\begin{align}
\text{AP@k}_u - \text{ AP@k для пользователя } u
\end{align}




### 0.5. DCG@k (Discount cumulative gain)

\begin{align}
\text{DCG@k} = \sum_{i=1}^{k} \frac{2^{rel_i} - 1}{\log_2(i+1)}
\end{align}

\begin{align}
\text{для бинарной релевантности это то же самое что}
\end{align}

\begin{align}
\text{DCG@k} = \sum_{i=1}^{k} \frac{rel_i}{\log_2(i+1)} = \text{rel}_1 + \sum_{i=2}^{k} \frac{rel_i}{\log_2(i+1)}
\end{align}


### 0.6. nDCG@k

\begin{align}
\text{nDCG@k} =  \frac{DCG@k}{IDCG@k}
\end{align}


Как соотносится nDCG@k с AP@k в случае бинарной релевантности?

### 0.7 MIUF - Mean Inverse User Frequency

\begin{align}
\text{MIUF@k для одного пользователя для одной выдачи рекомендаций:}
\end{align}
\begin{align}
\text{MIUF@k} = - \frac{1}{k} \sum_{i=1}^{k} {{\log_2{\frac{users(i)}{{n\_users}}}}}
\end{align}

\begin{align}
users(i) - \text{ это число пользователей, которые взаимодейтсвовали с }item_i
\end{align}

\begin{align}
n\_users - \text{число пользователей всего}
\end{align}

### 0.8 Serendipity

[Auralist: Introducing Serendipity into Music
Recommendation](https://researchswinger.org/publications/zhang12auralist.pdf)

\begin{align}
\text{для конкретного пользователя}
\end{align}

\begin{align}
Serendipity@k = \frac{1}{k} \sum_{i=1}^{k} max(p(i) - pu(i), 0) * rel(i)
\end{align}

\begin{align}
p(i) = \frac{(n\_items - i)}{n\_items} - \text{ вероятность рекомендовать item с рангом } i \text { данному пользователю}
\end{align}

\begin{align}
pu(i) = \frac{(n\_items - popularity_i)}{n\_items} - \text{ вероятность рекомендовать item с рангом } i \text { любому пользователю}
\end{align}


\begin{align}
rel_i - \text{бинарная релевантность item с рангом } i
\end{align}

\begin{align}
n\_items - \text{число items во всем каталоге}
\end{align}

\begin{align}
popularity_i - \text{ранг популярности items с рангом } i \text{ в выдаче}
\end{align}

### 0.9 Hit Rate

\begin{align}
HR@k = \frac{1}{n\_users} \sum_{i=1}^{n\_users} \text{ 1 если хотя бы одна рекомендация у пользователя из k попала верно, 0 } - \text{ иначе}
\end{align}

### 0.10 ARP@k

\begin{align}
\text{для одного пользователя для одной выдачи рекомендаций:}
\end{align}

\begin{align}
 ARP@k = \frac{1}{\left|U_{t}\right|}\sum_{u\in U_{t}^{}}\frac{\sum_{i\in L_{u}}\phi (i)}{\left| L_{u} \right |}
\end{align}

\begin{align}
\phi (i) -\text{ число предыдущих взаимодейтсвий с }  item _i.
\end{align}

\begin{align}
|U_{t}| -\text{ число пользователей в тестовом множестве }
\end{align}

\begin{align}
L_{u} -\text{ выдача из top-k рекомендаций по пользователю}
\end{align}

```python
import implicit
from scipy.sparse import coo_matrix
import numpy as np
```

```python
interactions_als = interactions.copy()
```

```python
interactions_als['weight'] = np.where(interactions_als['watched_pct'] > 10, 3, 1)
max_date = interactions_als['Datetime'].max()
# split dataset to train/test
train = interactions_als[interactions['Datetime'] < max_date - pd.Timedelta(days=7)].copy()
test = interactions_als[interactions['Datetime'] >= max_date - pd.Timedelta(days=7)].copy()
train.drop(train.query("total_dur < 300").index, inplace=True)
cold_users = set(test['user_id']) - set(train['user_id'])
test.drop(test[test['user_id'].isin(cold_users)].index, inplace=True)
cold_items = set(test['item_id']) - set(train['item_id'])
test.drop(test[test['item_id'].isin(cold_items)].index, inplace=True)
```

```python
# индексируем юзеров и айтемы от 0 до n-1
inverse_user_idx_dct = dict(enumerate(train['user_id'].unique()))
user_idx_dct = {v:k for k,v in inverse_user_idx_dct.items()}
inverse_item_idx_dct = dict(enumerate(train['item_id'].unique()))
item_idx_dct = {v:k for k,v in inverse_item_idx_dct.items()}
```

```python
train['user_idx'] = train['user_id'].map(user_idx_dct)
train['item_idx'] = train['item_id'].map(item_idx_dct)

test['user_idx'] = test['user_id'].map(user_idx_dct)
test['item_idx'] = test['item_id'].map(item_idx_dct)
```

```python
# Создаем матрицу интеракций: COO sparse matrix
row  = train['user_idx'].values
col  = train['item_idx'].values
data = train['weight'].values
matrix = coo_matrix((data, (row, col)))
```

```python
RANDOM_STATE = 42
num_threads = 1
ITERATIONS = 3
N_FACTORS = 128
K_RECOS = 10
ALPHA = 10
REG = 0.5
```

#### 1.1.1. Alternating Least Squares (ALS)

```python
als = implicit.als.AlternatingLeastSquares(factors = N_FACTORS, regularization = REG, alpha = ALPHA,
                                           iterations = ITERATIONS, random_state = RANDOM_STATE,
                                           num_threads=num_threads
                                           )
als.fit(matrix)
```

```python
%%time
test_users = test['user_idx'].unique()
preds, scores = als.recommend( userid = test_users,
                            user_items = matrix.tocsr()[test_users],
                              N = K_RECOS,
                              filter_already_liked_items=True)
```

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("data_original")
users = pd.read_csv(DATA_PATH / 'users.csv')
items = pd.read_csv(DATA_PATH / 'items.csv')
interactions = (
    pd.read_csv(DATA_PATH / 'interactions.csv', parse_dates=["last_watch_dt"])
    .rename(columns={"last_watch_dt": Columns.Datetime})
)
```

```python
# Process interactions
interactions[Columns.Weight] = np.where(interactions['watched_pct'] > 10, 3, 1)

# Split to train / test
max_date = interactions[Columns.Datetime].max()
train = interactions[interactions[Columns.Datetime] < max_date - pd.Timedelta(days=7)].copy()
test = interactions[interactions[Columns.Datetime] >= max_date - pd.Timedelta(days=7)].copy()
train.drop(train.query("total_dur < 300").index, inplace=True)
cold_users = set(test[Columns.User]) - set(train[Columns.User])
test.drop(test[test[Columns.User].isin(cold_users)].index, inplace=True)
test_users = test[Columns.User].unique()
catalog=train[Columns.Item].unique()

# Process user features to the form of a flatten dataframe
users.fillna('Unknown', inplace=True)
users = users.loc[users[Columns.User].isin(train[Columns.User])].copy()
user_features_frames = []
for feature in ["sex", "age", "income"]:
    feature_frame = users.reindex(columns=[Columns.User, feature])
    feature_frame.columns = ["id", "value"]
    feature_frame["feature"] = feature
    user_features_frames.append(feature_frame)
user_features = pd.concat(user_features_frames)

# Process item features to the form of a flatten dataframe
items = items.loc[items[Columns.Item].isin(train[Columns.Item])].copy()
items["genre"] = items["genres"].str.lower().str.replace(", ", ",", regex=False).str.split(",")
genre_feature = items[["item_id", "genre"]].explode("genre")
genre_feature.columns = ["id", "value"]
genre_feature["feature"] = "genre"
content_feature = items.reindex(columns=[Columns.Item, "content_type"])
content_feature.columns = ["id", "value"]
content_feature["feature"] = "content_type"
item_features = pd.concat((genre_feature, content_feature))
```

```python
# random.sample(seq, n)

dataset_no_features = Dataset.construct(
    interactions_df=train,
)

dataset_full_features = Dataset.construct(
    interactions_df=train,
    user_features_df=user_features,
    cat_user_features=["sex", "age", "income"],
    item_features_df=item_features,
    cat_item_features=["genre", "content_type"],
)

dataset_item_features = Dataset.construct(
    interactions_df=train,
    item_features_df=item_features,
    cat_item_features=["genre", "content_type"],
)

dataset_user_features = Dataset.construct(
    interactions_df=train,
    user_features_df=user_features,
    cat_user_features=["sex", "age", "income"],
)

feature_datasets = {
    "full_features": dataset_full_features,
    "item_features": dataset_item_features,
    "user_features": dataset_user_features
}
# Prepare dataset with biases as features

item_biases = pd.DataFrame({
    "id": catalog,
    "bias": 1
})
user_biases = pd.DataFrame({
    "id": train[Columns.User].unique(),
    "bias": 1
})

dataset_with_biases = Dataset.construct(
    interactions_df=train,
    user_features_df=user_biases,
    make_dense_user_features=True,
    item_features_df=item_biases,
    make_dense_item_features=True
)
feature_datasets["biases"] = dataset_with_biases
```

```python
metrics_name = {
    'MAP': MAP,
    'MIUF': MeanInvUserFreq,
    'Serendipity': Serendipity
}
metrics = {}
for metric_name, metric in metrics_name.items():
    for k in (1, 5, 10):
        metrics[f'{metric_name}@{k}'] = metric(k=k)
```

```python
RANDOM_STATE = 42
NUM_THREADS = -1
ITERATIONS = 3
N_FACTORS = 128
K_RECOS = 10
ALPHA = 10
REG = 0.5
```

## Biases

[Bias and Debias in Recommender System: A Survey and Future Directions (‘21)](https://arxiv.org/pdf/2010.03240.pdf)

Снимок экрана 2024-03-04 в 17.09.31.png



*   Popularity bias – чаще кликают / смотрят более популярные
*   Positivity bias – чаще кликают / смотрят более высокорейтинговые
*   Position bias – чаще кликают / смотрят на то что выше в выдаче
*   Exposure bias – товары одной группы / продавца суммарно показывают реже
*   User selection bias – пользователь склонен высоко оценивать понравившиеся ex. фильмы и не оценивать то что смотреть не стал (или оценивать минимальным рейтингом)
*   System selection bias – пользователь в основном взаимодействует с items, отобранные системой
*   Conformity Bias – а это вообще собирательный термин когда нам кажется что что-то не так и интеракции
пользователя в нашей системе с реальностью не совпадают
